Split data into 80/20

```python
from sklearn.model_selection import train_test_split

# Split with stratification to preserve class balance
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    stratify=y, 
    random_state=42
)

print("Train shape:", X_train.shape)
print("Test shape :", X_test.shape)
print("Train class balance:\n", y_train.value_counts())
```

Standardize the data using StandardScalar

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler only on training data
X_train_scaled = scaler.fit_transform(X_train)

# Use the same scaler to transform the test data
X_test_scaled = scaler.transform(X_test)

# Optional: Check shapes
print("Scaled X_train shape:", X_train_scaled.shape)
print("Scaled X_test shape :", X_test_scaled.shape)
```

Convert Scaled Data to DataFrame

```python
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
```

Corelation matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the Pearson correlation matrix
corr_matrix = X_train_scaled_df.corr()

# Plot heatmap with correlation values
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)
plt.title("Pearson Correlation Matrix (Scaled Features)")
plt.show()
```

Droping 10 variable with threshold of 0.9 to reduce multicolinarity

```python
import pandas as pd
import numpy as np

# Convert scaled array back to DataFrame with original feature names
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)

# Compute correlation matrix
corr_matrix = X_train_scaled_df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Set threshold and find features to drop
threshold = 0.9
to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

# Drop those features from both train and test sets (optional)
X_train_reduced = X_train_scaled_df.drop(columns=to_drop)

# Output
print("Dropped features due to high correlation (>", threshold, "):")
print(to_drop)
```

Droped same features from Test data as well

```python
# Convert test set to DataFrame with original feature names
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)

# Drop the same features that were dropped from training
X_test_reduced = X_test_scaled_df.drop(columns=to_drop)

# Optional: Check final shape
print("Shape of X_test_reduced:", X_test_reduced.shape)
```

Check for class inbalance

```python
from collections import Counter

# Check class distribution in training set
class_counts = Counter(y_train)
print("Class distribution in y_train:")
print(class_counts)
```

Use SMOTE to balance the class

```python
from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE on reduced and scaled training data
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_reduced, y_train)

# Check new class distribution
from collections import Counter
print("Class distribution after SMOTE:")
print(Counter(y_train_balanced))
```

Using SMOTE to balance the class without droping 10 variable

```python
from imblearn.over_sampling import SMOTE

# Initialize SMOTE
smote = SMOTE(random_state=42)

# Apply SMOTE on the full scaled training data (no feature reduction)
x_train_balanced_without_feature_reduction, y_train_balanced_without_feature_reduction = smote.fit_resample(
    X_train_scaled_df, y_train
)

# Check results (optional)
print("Balanced training shape:", x_train_balanced_without_feature_reduction.shape)
print("Balanced target distribution:\n", y_train_balanced_without_feature_reduction.value_counts())
```

Gridsearch to hyperparameter tuning

ROC Curve for All models

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize

# Binarize the labels (since ROC curve requires 0 and 1 labels)
y_test_bin = label_binarize(y_test, classes=[0, 1])

# Initialize the models' predictions
y_pred_prob_rf = grid_search_rf.best_estimator_.predict_proba(X_test_reduced)[:, 1]
y_pred_prob_mlp = grid_search_mlp.best_estimator_.predict_proba(X_test_reduced)[:, 1]
y_pred_prob_svc = svc_model.predict_proba(X_test_reduced)[:, 1]
y_pred_prob_dt = dt_model.predict_proba(X_test_reduced)[:, 1]
y_pred_prob_logreg = logreg.predict_proba(X_test_reduced)[:, 1]

# Compute ROC curve and AUC for each model
fpr_rf, tpr_rf, _ = roc_curve(y_test_bin, y_pred_prob_rf)
fpr_mlp, tpr_mlp, _ = roc_curve(y_test_bin, y_pred_prob_mlp)
fpr_svc, tpr_svc, _ = roc_curve(y_test_bin, y_pred_prob_svc)
fpr_dt, tpr_dt, _ = roc_curve(y_test_bin, y_pred_prob_dt)
fpr_logreg, tpr_logreg, _ = roc_curve(y_test_bin, y_pred_prob_logreg)

# Compute AUC for each model
roc_auc_rf = auc(fpr_rf, tpr_rf)
roc_auc_mlp = auc(fpr_mlp, tpr_mlp)
roc_auc_svc = auc(fpr_svc, tpr_svc)
roc_auc_dt = auc(fpr_dt, tpr_dt)
roc_auc_logreg = auc(fpr_logreg, tpr_logreg)

# Plot ROC curve for all models with distinct line styles and markers
plt.figure(figsize=(12, 10))

# Plot all ROC curves with different line styles and colors
plt.plot(fpr_rf, tpr_rf, color='blue', lw=2, linestyle='-', label=f'Random Forest (AUC = {roc_auc_rf:.2f})')
plt.plot(fpr_mlp, tpr_mlp, color='green', lw=2, linestyle='--', label=f'Neural Network (AUC = {roc_auc_mlp:.2f})')
plt.plot(fpr_svc, tpr_svc, color='red', lw=2, linestyle=':', label=f'SVC (AUC = {roc_auc_svc:.2f})')
plt.plot(fpr_dt, tpr_dt, color='orange', lw=2, linestyle='-.', label=f'Decision Tree (AUC = {roc_auc_dt:.2f})')
plt.plot(fpr_logreg, tpr_logreg, color='purple', lw=2, linestyle='-', label=f'Logistic Regression (AUC = {roc_auc_logreg:.2f})')

# Plot the diagonal (no skill, random classifier)
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=2)

# Labels and title
plt.title('ROC Curve for All Models', fontsize=16)
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate (Recall)', fontsize=14)
plt.legend(loc='lower right', fontsize=12)
plt.tight_layout()

# Show plot
plt.show()
```
